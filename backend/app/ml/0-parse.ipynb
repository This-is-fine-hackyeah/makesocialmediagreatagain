```python
import sys
sys.path.append("..")
import json
from app.core.process.extract import download_media, extract_text
```

```python
with open("negative.json") as fp:
    negative = json.load(fp)

download_media(negative, destination="negative_images")
```

```python
# positive must be copied from the provided dataset
# and put in `positive_images`
```

```python
extract_text("negative_images")
extract_text("positive_images")
```

```python
# final parse negative
with open("negative_images.json") as fp:
    negative_images = json.load(fp)
negative_parsed = {
    v["id"]: [v["text"].strip(), negative_images.get(v["id"], "").strip()]
    for v in negative
}
negative_parsed = {
    k: " ".join(v).strip()
    for k, v in negative_parsed.items()
}

with open("negative_parsed.json", "w") as fp:
    json.dump(negative_parsed, fp)
```

```python
# final parse positive
with open("positive_images.json") as fp:
    positive_images = json.load(fp)
positive_parsed = {
    k: v[0].strip()
    for k, v in positive_images.items()
}
positive_parsed.pop(".gitkeep")

with open("positive_parsed.json", "w") as fp:
    json.dump(positive_parsed, fp)
```
